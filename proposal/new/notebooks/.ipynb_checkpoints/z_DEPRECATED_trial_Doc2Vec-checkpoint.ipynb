## 以下是 Word2Vec 部分

```python
# Step 0: Import modules

import gensim
import os
import numpy as np
import pandas as pd
import random
import pandas as pd
import numpy as np
from IPython.display import display
%matplotlib inline

print "Step 0 succeed."
```

```python
text8str = ""
with open(os.path.join(os.getcwd(), 'text8'), 'r') as readf:
    text8str = readf.read()

text8 = text8str.split()#gensim.corpora.dictionary.Dictionary([text8str.split()])
print len(text8)

sentences = gensim.models.word2vec.Text8Corpus('text8')
model = gensim.models.word2vec.Word2Vec(sentences=sentences, iter=10, size=200, workers=4, min_count=1)

print "Model created. See some test:"

print model.wv.most_similar(positive=['king', 'woman'], topn=1)
```

```python
# save model
model.save('gensimModel.word2vec')
```

```python
# Step 1: preprocess the data
paths = {}
paths['dir.train'] = os.path.join(os.getcwd(), 'trialdata', 'train')
paths['dir.test'] = os.path.join(os.getcwd(), 'trialdata', 'test')

stopwords = ""
with open('stoplist.txt', 'r') as stoplistfile:
    stopwords = stoplistfile.read()
stopwords = stopwords.split()
with open('stoplist2.txt', 'r') as stoplistfile2:
    stopwords2 = stoplistfile2.read()
    stopwords2 = stopwords2.split('\n')
    stopwords = set(stopwords)
    stopwords = list(stopwords.union(set(stopwords)))

vocabList = model.wv.vocab.items()[:200]
    
print "Step 1 Succeed"
```

```python
# Step 2: read data and save it in data['vec.train'] 和 data['vec.test']

data = {}
data['vec.train'] = {'w2v.mean':[], 'class':[]}
data['vec.test'] = {'w2v.mean':[], 'class':[]}

for tpart in ['train', 'test']:
    dirpath = paths['dir.{}'.format(tpart)]
    for (ind, cls) in enumerate(os.listdir(dirpath)):
        clspath = os.path.join(dirpath, cls)
        files = os.listdir(clspath)
        for f in files:
            fpath = os.path.join(clspath, f)
            with open(fpath, 'r') as readf:
#                 print "new file"
                tokens = [token for token in readf.read().split() if token not in stopwords]#readf.read().split()#
                # Word2Vec representation
                # begin
                vec = np.array([0.0 for i in range(200)])
                expectationVal = np.array([0.0 for i in range(200)])
                for token in tokens:
                    try:
    #                     print "token: {}".format(token)
    #                     print model[token]
                        vec = model[token]
                    except:
                        vec += expectationVal
#                 print "before division: {}".format(vec)
                vec = vec / float(len(tokens))
#                 print "after division: {}".format(vec)
                 # end
            data['vec.{}'.format(tpart)]['w2v.mean'].append(vec)
            data['vec.{}'.format(tpart)]['class'].append(cls)
#             print "\n\n"
    tmp = data['vec.{}'.format(tpart)]
    ind = (random.sample(range(len(tmp['class'])), 1))[0]
    print "sample(transformed) from {}[{}]:\n[corpus]\n {}\n[class]\n{}".format(tpart, ind, tmp['w2v.mean'][ind], tmp['class'][ind])
    print

print "Step 2 Succeed"
```

```python
# tmpa = np.array([0.0 for i in range(200)])
# tmpa += model['gatech'] 
# print tmpa
# # print model['gatech']

# initialization should be the same type as elements that will be used to add
```

```python
# Step 3: Save in Pandas.DataFrame
#
# 将 data['matrix.train'] 与 data['matrix.test'] 转换成 Pandas.DataFrame 格式，保存到 df['train'] 和 df['test'] 中（df 为字典格式：String -> DataFrame）

df = {}
for tpart in ['train', 'test']:
    datadict = {}
    datadict['class'] = data['vec.{}'.format(tpart)]['class']
    datavec = np.array(data['vec.{}'.format(tpart)]['w2v.mean'])
    for col in range(200):
        datadict[col]= datavec[:, col]

    df[tpart] = pd.DataFrame(data=datadict)
    print "See df[{}]".format(tpart)
    display(df[tpart])
    print "\n\n\n"

print "Step 3 Succeed."

# 繁琐点：研究如何把 CSR 矩阵中的数据规整好放到 DataFrame 中，并与 Class 一一对应
```

```python
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.neural_network import MLPClassifier

X_train = df['train'].drop('class', axis=1)
y_train = df['train']['class']

clf = KNeighborsClassifier()
# clf = LinearSVC()
# clf = SGDClassifier()
# clf = GaussianNB()
# clf = DecisionTreeClassifier()
# clf =MultinomialNB()
# clf = MLPClassifier()
clf.fit(X_train, y_train)
# print "Training fishied with clf with [n_classes, n_features]: {}".format(clf.coef_)
print "Step 4 finished"
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

X_test = df['test'].drop('class', axis=1)
y_test_true = df['test']['class']

y_test_pred = clf.predict(X_test)
print accuracy_score(y_test_true, y_test_pred)
print f1_score(y_test_true, y_test_pred, average='macro')
print f1_score(y_test_true, y_test_pred, average='micro')
```

## 以下是 Doc2Vec 部分

```python
# Step 0: Import modules
import gensim
import os
import numpy as np
import pandas as pd
```

```python
# Step 1: preprocess the data
paths = {}
paths['dir.train'] = os.path.join(os.getcwd(), 'trialdata', 'train')
paths['dir.test'] = os.path.join(os.getcwd(), 'trialdata', 'test')

print "Step 1 Succeed"
```

```python
# Step 2: read data and save it in data['nearRaw.train'] 和 data['nearRaw.test']

## Step 2.A: get stopwords

import random
data = {}
data['nearRaw.train'] = {'corpus':[], 'class':[]}
data['nearRaw.test'] = {'corpus':[], 'class':[]}
stopwords = ""
with open('stoplist.txt', 'r') as stoplistfile:
    stopwords = stoplistfile.read()
stopwords = stopwords.split()
with open('stoplist2.txt', 'r') as stoplistfile2:
    stopwords2 = stoplistfile2.read()
    stopwords2 = stopwords2.split('\n')
    stopwords = set(stopwords)
    stopwords = list(stopwords.union(set(stopwords2)))
    
print "Step 2.A sucdeed."
```

```python
print 'to' in stopwords
```

```python
## Step 2.B: read and save data
    
for tpart in ['train', 'test']:
    dirpath = paths['dir.{}'.format(tpart)]
    for (ind, cls) in enumerate(os.listdir(dirpath)):
        clspath = os.path.join(dirpath, cls)
        files = os.listdir(clspath)
        for f in files:
            fpath = os.path.join(clspath, f)
            with open(fpath, 'r') as readf:
                tokens = [token for token in readf.read().split() if token not in stopwords]#readf.read().split()#
                if 'train' == tpart:
                    corpus = gensim.models.doc2vec.TaggedDocument(tokens, [cls])
                else:
                    corpus = tokens
                data['nearRaw.{}'.format(tpart)]['corpus'].append(corpus)
                data['nearRaw.{}'.format(tpart)]['class'].append(cls)
    tmp = data['nearRaw.{}'.format(tpart)]
    ind = (random.sample(range(len(tmp['class'])), 1))[0]
    print "sample(transformed) from {}[{}]:\n[corpus]\n {}\n[class]\n{}".format(tpart, ind, tmp['corpus'][ind], tmp['class'][ind])
    print
    
print "Step 2.B Succeed"
print "Step 2 Succeed"
```

```python
# Step 3: Training the Model
## Substep A: Initialization
vocabularySize = 70
model = gensim.models.doc2vec.Doc2Vec(size=vocabularySize, iter=250, workers=4, alpha=0.025, min_alpha=0.005)
print "Substep A succeed."
```

```python
## Substep B: Build vocabulary from corpus
model.build_vocab(data['nearRaw.train']['corpus'])
# text8str = ""
# with open(os.path.join(os.getcwd(), 'text8'), 'r') as readf:
# text8str = readf.read()

# # text8 = gensim.corpora.dictionary.Dictionary([text8str.split()])
# print "text8 created"
# model.build_vocab([text8])
print "Substep B succeed."
# print model
```

```python
# Substep D: Train models
%time model.train(sentences=data['nearRaw.train']['corpus'], total_examples=model.corpus_count)
```

```python
print data['nearRaw.test']['corpus'][25]
```

```python
# Substep E: Infer vectors for posts
for tpart in ['train', 'test']:
    countCorpus = len(data['nearRaw.{}'.format(tpart)]['corpus'])
    data['matrix.{}'.format(tpart)] = []
    for ind in range(countCorpus):
        if tpart == 'train':
            data['matrix.{}'.format(tpart)].append(model.infer_vector(data['nearRaw.{}'.format(tpart)]['corpus'][ind][0]))
        else:
            data['matrix.{}'.format(tpart)].append(model.infer_vector(data['nearRaw.{}'.format(tpart)]['corpus'][ind]))
    sampleind = (random.sample(range(countCorpus), 1))[0]
    print "sample for matrix.{}".format(tpart)
    print "from ind: {}".format(sampleind)
    print data['matrix.{}'.format(tpart)][sampleind]

print "Substep C succeed"
```

```python
## Step 4: save into DataFrame
from IPython.display import display
%matplotlib inline
df = {}
for tpart in ['train', 'test']:
    datadict = {}
    datadict['class'] = data['nearRaw.{}'.format(tpart)]['class']
    for feature_ind in range(vocabularySize):#range(len(data['matrix.{}'.format(tpart)][0])):#
        m = np.array(data['matrix.{}'.format(tpart)])
        datadict[feature_ind] = m[:,feature_ind]

    df[tpart] = pd.DataFrame(data=datadict)
    print "See df[{}]".format(tpart)
    display(df[tpart])

print "Step 4 succeed."
```

```python
varFeature = []
for i in range(vocabularySize):
    varFeature.append( np.var(df['train'][i]))

indices = []
varTop = sorted(varFeature)
indices = [varFeature.index(i) for i in varTop]
#indices.reverse()
```

```python
# Step 5: Classifier learning
from sklearn.svm import LinearSVC
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
featuresDrop = ['class'] + indices[:20]
X_train = df['train'].drop(featuresDrop, axis=1)
y_train = df['train']['class']
# clf = LinearSVC()
clf = SVC(kernel='linear')
# clf = GaussianNB()
clf.fit(X_train, y_train)
print "Step 5 finished"
```

```python
# Step 6: evaluation
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
X_test = df['test'].drop(featuresDrop, axis=1)
y_test_true = df['test']['class']
y_test_pred = clf.predict(X_test)
print accuracy_score(y_test_true, y_test_pred)
print f1_score(y_test_true, y_test_pred, average='macro')
print f1_score(y_test_true, y_test_pred, average='micro')
```
